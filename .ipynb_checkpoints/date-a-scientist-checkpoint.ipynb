We are still removing a lot of the data, about 20%, but at this point I think 47791 rows are what we are going to stick with as I want to leave in the variable 'job' as a predictor value; job has 8198 NA values.

Now to figure out how to encode the categorical varibles, as almost all of them are, into numerical variables. There are a few methods to doing this:
1). Label Encoding: Gives sequential weights to categorical varibles, but then the model is stuck with weights that could not make much sense. For example if using the 'status' variable, 'single' could be 1, 'available' = 2, 'seeing someone' = 3, etc, which does not make a lot of sense.
2). One Hot Encoding: Changes the different answers for the variables into columns and puts 1's and 0's for each.
3). Custom Binary Encoding: Combination of label encoding and one hot encoding.

I am going to use one hot encoding and both pandas and scikit learn have this functionality. To do this, I am going to create a dataframe with only the predictors I will use and the outcome variable. One hot encoding is going to add a whole lot of columns, so I would like to keep this as condensed as possible. I am going to perform one hot encoding on 'body_type', 'job', 'orientation', and 'status'.

```python
profiles = okc_no_na[['body_type', 'height', 'income', 'job', 'orientation', 'sex', 'status']]
#print(profiles.head())
print(profiles.dtypes)
```

```python
# from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
ohe_results = ohe.fit_transform(profiles[['body_type']])
ohe_1 = pd.DataFrame(ohe_results.toarray(), columns=ohe.categories_)
#print(ohe_1.head())
#print(len(ohe_1))
ohe_results_2 = ohe.fit_transform(profiles[['job']])
ohe_2 = pd.DataFrame(ohe_results_2.toarray(), columns=ohe.categories_)
#print(ohe_2.head())
ohe_results_3 = ohe.fit_transform(profiles[['orientation']])
ohe_3 = pd.DataFrame(ohe_results_3.toarray(), columns=ohe.categories_)
# print(ohe_3.head())
# print(len(ohe_3))
# ohe_results_4 = ohe.fit_transform(profiles[['status']])
# ohe_4 = pd.DataFrame(ohe_results_4.toarray(), columns=ohe.categories_)

ml_ready = profiles[['height', 'sex']]
#print(ml_ready.head())
#len(ml_ready)
#print(len(ml_ready))
#Have to reset the index to keep concat from inserting index values where the indexes do not match
ml_ready.reset_index(drop=True, inplace=True)
ohe_1.reset_index(drop=True, inplace=True)
ohe_2.reset_index(drop=True, inplace=True)
ohe_3.reset_index(drop=True, inplace=True)
# ohe_4.reset_index(drop=True, inplace=True)
ml_ready = pd.concat([ml_ready, ohe_1, ohe_2, ohe_3], axis=1)
# print(ml_ready.head())
print(len(ml_ready))
ml_ready.isna().sum().sum()
```

```python
ml_ready['sex'] = ml_ready['sex'].map({'m': 0,'f': 1})
```

```python
print(ml_ready.sex.value_counts())
```

# Building the model

```python
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier

labels = ml_ready.sex
ml_ready = ml_ready.drop('sex', axis=1)
train, test, train_labels, test_labels = train_test_split(ml_ready, labels, random_state = 1)
print(len(train), len(test))

forest = RandomForestClassifier(random_state=1)
forest.fit(train, train_labels)
print(forest.score(test, test_labels))
```

I am removing some of the predictors to see if I can get a higher accuracy. I think that removing variables will lead to less over fitting in the long run.

Accuracy for: 'body_type', 'height', 'income', 'job', 'orientation', 'sex', and 'status' = 0.867
Accuracy for: 'body_type', 'height', 'income', 'job', 'orientation', and 'sex' = 0.866
Accuracy for: 'body_type', 'height', 'job', 'orientation', and 'sex' = 0.869

It seems like if I remove many of these different variables, the accuracy does not move much. I do not know if I should remove all that have little bearing on accuracy or if I should leave them in. for now, I am going to leave them in and go with the last model listed above.
